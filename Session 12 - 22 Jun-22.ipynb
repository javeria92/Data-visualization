```python
fig, ax = plt.subplots(figsize = (4.5,6)) # canvas -> oo interface

ax.barh(deaths["Country_Other"], deaths["Total_Deaths"], height = 0.45, color = "#af0b1e") 

for spine in ["left", "right", "bottom", "top"]:
    ax.spines[spine].set_visible(False)
    
ax.tick_params(left =0,bottom =0,)

ax.set_xticks([0,150000,300000])

ax.xaxis.tick_top()

ax.tick_params(top = 0, axis = "x" , colors = "grey")

ax.text(-80000, y = 23.5, s = "The Death Toll Wordlwide", weight = "bold", size = 17)
ax.text(-80000, y = 22.3, s = "Top 20 Countries by death toll", size = 12)

# ax.set_xticklabels(["0", "150,000", "300,000"])

ax.set_xticklabels(["0", "150,000", "300,000"])

ax.set_yticklabels([])


countries = deaths["Country_Other"]

for i, c in zip(range(20),countries):
    ax.text(-80000, i-0.15, s = c)
    
ax.axvline(x = 150000, ymin = 0.045, c = "grey", alpha = 0.4)

plt.show()
```
