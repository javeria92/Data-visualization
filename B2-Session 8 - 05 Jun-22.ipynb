```python
traffic_per_day = {}

for i, day in zip(range(0,135,27),days):
    # print(i,i+27, day)
    traffic_per_day[day] = traffic[i:i+27]
```

```python
traffic_per_day["Monday"].shape
```

```python
traffic_per_day["Monday"].plot.line(x = "Hour (Coded)", y = "Slowness in traffic (%)")
plt.show()
```

```python
for day in days:
    traffic_per_day[day].plot.line(x = "Hour (Coded)", y = "Slowness in traffic (%)")
    plt.title(day)
    plt.ylim(0,25)
plt.show()
```

```python
for day in days:
    plt.plot(traffic_per_day[day]["Hour (Coded)"], traffic_per_day[day]["Slowness in traffic (%)"])
plt.show()
```

```python
plt.figure(figsize = (10,12))

for i, day in zip(range(1,6), days):
    plt.subplot(3,2,i)
    plt.plot(traffic_per_day[day]["Hour (Coded)"], traffic_per_day[day]["Slowness in traffic (%)"])
    plt.title(day)
    plt.ylim(0,25)
    plt.xlim(0,28)
    
plt.subplot(3,2,6)
for day in days:
    plt.plot(traffic_per_day[day]["Hour (Coded)"], traffic_per_day[day]["Slowness in traffic (%)"], label = day)
    plt.legend()
plt.savefig("grid.png")
plt.show() 
```
