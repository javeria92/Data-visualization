```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
titanic = pd.read_csv("titan.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic = titanic.dropna()
```

```python
titanic.shape
```

```python
sns.histplot(titanic["Age"])

plt.show()
```

```python
sns.displot(titanic["Age"])

plt.show()
```

```python
sns.distplot(titanic["Age"])

plt.show()
```

```python
sns.kdeplot(titanic["Age"])

plt.show()
```

```python
# progress or regress
```

```python
sns.kdeplot(titanic["Age"], shade = True)

plt.show()
```

```python
# sns.set_theme()

sns.set_style("ticks") # white, whitegrid, darkgrid, dark
```

```python
sns.kdeplot(titanic["Age"], shade = True)

sns.set_style("white")

plt.show()
```

```python
sns.kdeplot(titanic["Age"], shade = True)

sns.set_style("whitegrid")

plt.show()
```

```python
sns.set_style("darkgrid")

sns.kdeplot(titanic["Age"], shade = True)

plt.show()
```

```python
sns.set_style("dark")

sns.kdeplot(titanic["Age"], shade = True)

plt.show()
```

```python
sns.set_theme()

sns.kdeplot(titanic["Age"], shade = True)

plt.show()
```

```python
sns.set_style("white")

sns.kdeplot(titanic["Age"], shade = True)

plt.show()
```

```python
sns.set_style("white")

sns.kdeplot(titanic["Age"], shade = True)
sns.despine()

plt.show()
```

```python
sns.set_style("white")

sns.kdeplot(titanic["Age"], shade = True)
sns.despine(left = True, bottom = True)

plt.show()
```

```python
g = sns.FacetGrid(titanic, col = "Survived", height = 3) # (1,2)

g.map(sns.kdeplot, "Age", shade = True)

plt.show()
```

```python
g = sns.FacetGrid(titanic, col = "Survived", height = 3) # (1,2)

g.map(plt.hist, "Age")

plt.show()
```

```python
g = sns.FacetGrid(titanic, col = "Survived", height = 3) # (1,2)

g.map(sns.boxplot, "Age")

plt.show()
```

```python
sns.boxplot(data = titanic , x = "Survived", y = "Age")
plt.show()
```

```python
sns.FacetGrid(titanic, col = "Survived", row = "Pclass", height = 3).map(sns.kdeplot, "Age", shade = True) # (3,2)

plt.show()
```

```python
# pip install seaborn
```

```python
g = sns.FacetGrid(titanic, col = "Survived", row = "Pclass", hue = "Sex")
g.map(sns.kdeplot, "Age", shade = True,warn_singular=False) # (3,2)

g.add_legend()
sns.despine(left = True, bottom = True)

plt.ylim(0,0.04)

plt.show()
```
