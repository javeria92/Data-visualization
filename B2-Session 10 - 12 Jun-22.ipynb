```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
degrees = pd.read_csv("degrees.csv")
```

```python
maj_cat = ['Biology', 'Computer Science', 'Engineering', 'Math and Statistics']

fig = plt.figure(figsize = (12,12))


for i in range(4):
    ax = fig.add_subplot(2,2,i+1)
    ax.plot(degrees["Year"], degrees[maj_cat[i]], label = "Women")
    ax.plot(degrees["Year"], 100-degrees[maj_cat[i]], label = "Men")
    ax.set_xlim(1968,2012)
    ax.set_ylim(0,100)
    
    ax.set_title(maj_cat[i])


    ax.tick_params(left = False, bottom = False)

    for v in ax.spines.values():
        v.set_visible(False)

ax.legend(loc = "upper left")

plt.show()
```
